```python
from selenium import webdriver
from selenium.webdriver.common.by import By 
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import Select
from selenium.common.exceptions import NoSuchElementException, ElementClickInterceptedException, ElementNotInteractableException

import time

driver=webdriver.Chrome(service=Service(ChromeDriverManager().install()))
machine_details=[]
driver.get('https://www.amazon.in/s?k=washing+machine')

for i in range(1,20):
    
    time.sleep(3)
    driver.maximize_window()
    products=driver.find_elements(By.CSS_SELECTOR,'.a-section.a-spacing-none.puis-padding-right-small.s-title-instructions-style')

    for product in products:
        time.sleep(3)
        pro=product.find_element(By.CSS_SELECTOR,'.a-link-normal.s-underline-text.s-underline-link-text.s-link-style.a-text-normal').click()
        time.sleep(3)
        p = driver.current_window_handle

        parent = driver.window_handles[0]

        chld = driver.window_handles[1]

        driver.switch_to.window(chld)


        try:
            name=driver.find_element(By.XPATH,'//*[@id="productTitle"]').text
        except NoSuchElementException:
            name='NaN'
            pass
        try:    
            stars = driver.find_element(By.ID,'acrPopover').get_attribute('title')
        except NoSuchElementException:
            stars='NaN'
            pass

        try:
            brand=driver.find_element(By.XPATH,"//tr[@class='a-spacing-small po-brand']//td[@class='a-span9']").text
        except NoSuchElementException:
            brand="NaN"
            pass
        try:
            price=driver.find_element(By.CSS_SELECTOR,'#corePriceDisplay_desktop_feature_div > div.a-section.a-spacing-none.aok-align-center > span.a-price.aok-align-center.reinventPricePriceToPayMargin.priceToPay > span:nth-child(2) > span.a-price-whole').text
        except NoSuchElementException:
            price="NaN"
            pass
        try:
            capacity=driver.find_element(By.CSS_SELECTOR,'.a-spacing-small.po-capacity').text
        except NoSuchElementException:
            capacity="NaN"
            pass

        try:
            aloc=driver.find_element(By.CSS_SELECTOR,'.a-spacing-small.po-access_location').text
            if(aloc==""):
                aloc="NaN"
        except NoSuchElementException:
            aloc="NaN"
            pass
        machine_details.append({'Name':name,
                                'Review stars':stars,
                                'Brand':brand,
                                'Price':price,
                                'Capacity':capacity,
                                'Access Location':aloc})
        
        products=[]
        driver.close()
        driver.switch_to.window(parent)
    npg=driver.find_element(By.CSS_SELECTOR,'.s-pagination-item.s-pagination-next.s-pagination-button.s-pagination-separator').click()    

df=pd.DataFrame(machine_details)
df
```
